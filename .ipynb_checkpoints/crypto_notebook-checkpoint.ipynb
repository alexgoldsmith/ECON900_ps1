```python
#Load libraries
import pandas as pd
import numpy as np
```

```python
#Read data to DataFrame
df = pd.read_json('currencies.json')
```

```python
#Let's get a sense of the data
df.head()
```

```python
#DataFrame structure
df.info()
```

```python
#Clean market rank variable
df['market_rank'] = (
df['market_rank']
    .str.replace('#','')
    .astype(int)
)

#Clean day close variable
df['day_close'] =  (
df['day_close']
    .str.replace(',','')
    .astype(float)
)
#Check results
df.head()
```

```python
#Sort DataFrame
df.sort_values(['long_name', 'date'], inplace = True)
df.head()
```

```python
#Take first differnces
df['close_diff'] = df.groupby(['long_name'])['day_close'].transform(lambda x: x.diff())
df.head()
```

```python
#Create example df
bitcoin_df = df.loc[df['long_name'] == 'Bitcoin']
bitcoin_df.head()
```
